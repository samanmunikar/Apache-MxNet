```python
import mxnet as mx
import gluoncv as gcv
import matplotlib.pyplot as plt
```

Download the image using gcv.utils.download

```python
image_url="https://raw.githubusercontent.com/dmlc/web-data/master/gluoncv/classification/mt_baker.jpg"
image_filepath="mt_baker.jpg"
gcv.utils.download(url=image_url, path=image_filepath)
```

```python
# Load Image

image = mx.image.imread('mt_baker.jpg')
print("Image shape", image.shape) # Height X Width X Channel shape
print("Image dtype", image.dtype) # Unsigned 8 bit Integers
print("Image min", image.min().asscalar()) # Minimum pixel value in the image
print("Image max", image.max().asscalar()) # Maximum pixel value in the image
```

```python
# Visualize Image

plt.imshow(image.asnumpy())
```

```python
# Transform our image as the training image were trained
# so that the model can understand our image

image = gcv.data.transforms.presets.imagenet.transform_eval(image)
print("Image shape", image.shape) # Batch, Channel, Height, Width
print("Image dtype", image.dtype) # 32 bit Float
print("Image min", image.min().asscalar()) # Normalized as per ImageNet1K Statistics with mean 0 and std 1
print("Image max", image.max().asscalar()) # Normalized as per ImageNet1K Statistics with mean 0 and std 1
```

```python
# Load Model from model zoo

network = gcv.model_zoo.get_model('ResNet50_v1d', pretrained=True)
```

```python
# Make Prediction

prediction = network(image)
print(type(prediction))
print(prediction.shape) # Batch, prediction(confidence) of our image on each of 1000 classes
```

```python
prediction = prediction[0] # Remove the batch dimension because we have only one batch of image
```

```python
# Convert logits into probability

probability = mx.nd.softmax(prediction)
```

```python
# Get topk probability
k=5
topk_indicies = mx.nd.topk(probability, k=k)
topk_indicies
```

```python
# Get the class label of that indicies
network.classes[980]
```

```python
#Display topk labels for the image
for i in range(k):
    class_index = topk_indicies[i].astype('int').asscalar()
    class_label = network.classes[class_index]
    class_probability = probability[class_index]
    print("#{} {} {:.3f}%".format(i+1, class_label, class_probability.asscalar()*100))
    
```
